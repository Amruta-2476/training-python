```python
# queue without size
# enqueue, dequeue, isEmpty, peek(front), isFull, delete, display, exit
class Queue:
    def __init__(self):
        self.myqueue = []

    def enqueue(self, val):
        self.myqueue.append(val)
        print(f'Added {val} in queue.')

    def dequeue(self):
        if self.isEmpty():
            print('Queue empty. Cannot pop')
        else:
            dequeue_val = self.myqueue.pop(0)
            print(f'Removed {dequeue_val} from queue.')

    def peek(self):
        if self.isEmpty():
            print('Queue empty. Cannot remove element')
        else:
            print(f'Element at front is {self.myqueue[0]}')

    def show(self):
        print(self.myqueue,' is the queue.')

    def isEmpty(self):
        return len(self.myqueue) == 0
    
    def delete(self):
        self.myqueue.clear()
        print(self.myqueue)
        

ob = Queue()
while True:
    ch = int(input('enter choice: '))
    print('1. Enqueue')
    print('2. Dequeue')
    print('3. Peek')
    print('4. isEmpty')
    print('5. Show')
    print('6. Delete')
    print('7. Exit')

    if ch == 1:
        val = int(input('enter value to push: '))
        ob.enqueue(val)
    elif ch == 2:
        ob.dequeue()
    elif ch == 3:
        ob.peek()
    elif ch == 4:
        print(ob.isEmpty())
    elif ch == 5:
        ob.show()
    elif ch == 6:
        ob.delete()
    elif ch == 7:
        print('exiting . . .')            
        break
    else:
        print('invalid choice')
```

```python
# queue with size
# enqueue, dequeue, isEmpty, peek(front), isFull, delete, display, exit
class Queue:
    def __init__(self, max):
        self.myqueue = []
        self.max = max

    def enqueue(self, val):
        if self.isFull():
            print('Queue full. cannot add')
        else:
            self.myqueue.append(val)
            print(f'Added {val} in queue.')

    def dequeue(self):
        if self.isEmpty():
            print('Queue empty. Cannot pop')
        else:
            dequeue_val = self.myqueue.pop(0)
            print(f'Removed {dequeue_val} from queue.')

    def peek(self):
        if self.isEmpty():
            print('Queue empty. Cannot remove element')
        else:
            print(f'Element at front is {self.myqueue[0]}')

    def show(self):
        print(self.myqueue,' is the queue.')

    def isEmpty(self):
        return len(self.myqueue) == 0
    
    def isFull(self):
        return len(self.myqueue) == max
    
    def delete(self):
        self.myqueue.clear()
        print(self.myqueue)
        

max_size = int(input('enter size of queue: '))
ob = Queue(max_size)
while True:
    ch = int(input('enter choice: '))
    print('1. Enqueue')
    print('2. Dequeue')
    print('3. Peek')
    print('4. isEmpty')
    print('5. Show')
    print('6. Delete')
    print('7. isFull')
    print('8. Exit')

    if ch == 1:
        val = int(input('enter value to push: '))
        ob.enqueue(val)
    elif ch == 2:
        ob.dequeue()
    elif ch == 3:
        ob.peek()
    elif ch == 4:
        print(ob.isEmpty())
    elif ch == 5:
        ob.show()
    elif ch == 6:
        ob.delete()
    elif ch == 7:
        ob.isFull()
    elif ch == 8:
        print('exiting . . .')            
        break
    else:
        print('invalid choice')
```
