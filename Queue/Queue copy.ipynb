```python
class Queue:
    def __init__(self):
        self.queue = []
        print("Queue created successfully!")

    def enqueue(self, x):
        self.queue.append(x)
        print(f'Enqueued: {x}')

    def dequeue(self):
        if self.isEmpty():
            print('Queue is empty. Cannot dequeue.')
            return None
        else:
            dequeued_value = self.queue.pop(0)   # ******* important ***********
            # dequeued_value = self.queue.remove(self.queue[0])     this gives dequeued_value = None
            print(f'Dequeued: {dequeued_value}')
            return dequeued_value

    def front(self): # peek
        if self.isEmpty():
            print('Queue is empty. No front element.')
            return None
        else:
            print(f'Front element: {self.queue[0]}')
            return self.queue[0]

    def isEmpty(self):
        return len(self.queue) == 0

    def show(self):
        print('Queue:', self.queue)

    def delete(self):
        self.queue.clear()
        print('Queue deleted successfully.')

# Main menu for queue operations
obj = Queue()  # Creating queue object

while True:
    print('\n1. Enqueue')
    print('2. Dequeue')
    print('3. Show Queue')
    print('4. Check if Queue is Empty')
    print('5. View Front Element')
    print('6. Delete Queue')
    print('7. Exit')

    choice = int(input('Enter option: '))

    if choice == 1:
        value = int(input('Enter value to enqueue: '))
        obj.enqueue(value)
    elif choice == 2:
        obj.dequeue()
    elif choice == 3:
        obj.show()
    elif choice == 4:
        print('Queue is empty.' if obj.isEmpty() else 'Queue is not empty.')
    elif choice == 5:
        obj.front()
    elif choice == 6:
        obj.delete()
    elif choice == 7:
        print('Exiting...')
        break
    else:
        print('Invalid choice. Please enter a valid option.')
```

```python
# Queue Implementation (With Size Limit)

class Queue:
    def __init__(self, max_size):
        self.queue = []
        self.max_size = max_size  # Set the maximum size of the queue
        print(f"Queue created successfully with size limit {max_size}!")

    def enqueue(self, x):
        if self.isFull():
            print("Queue is full. Cannot enqueue.")
        else:
            self.queue.append(x)
            print(f"Enqueued: {x}")

    def dequeue(self):
        if self.isEmpty():
            print("Queue is empty. Cannot dequeue.")
            return None
        else:
            dequeued_value = self.queue.pop(0)
            print(f"Dequeued: {dequeued_value}")
            return dequeued_value

    def front(self):
        if self.isEmpty():
            print("Queue is empty. No front element.")
            return None
        else:
            print(f"Front element: {self.queue[0]}")
            return self.queue[0]

    def isEmpty(self):
        return len(self.queue) == 0

    def isFull(self):
        return len(self.queue) >= self.max_size

    def show(self):
        print("Queue:", self.queue)

    def delete(self):
        self.queue.clear()
        print("Queue deleted successfully.")

# Get user-defined queue size
max_size = int(input("Enter the maximum queue size: "))
obj = Queue(max_size)  # Creating a queue with a size limit

# Main menu for queue operations
while True:
    print("\n1. Enqueue")
    print("2. Dequeue")
    print("3. Show Queue")
    print("4. Check if Queue is Empty")
    print("5. Check if Queue is Full")
    print("6. View Front Element")
    print("7. Delete Queue")
    print("8. Exit")

    choice = int(input("Enter option: "))

    if choice == 1:
        if obj.isFull():
            print("Cannot enqueue, queue is full!")
        else:
            value = int(input("Enter value to enqueue: "))
            obj.enqueue(value)
    elif choice == 2:
        obj.dequeue()
    elif choice == 3:
        obj.show()
    elif choice == 4:
        print("Queue is empty." if obj.isEmpty() else "Queue is not empty.")
    elif choice == 5:
        print("Queue is full." if obj.isFull() else "Queue is not full.")
    elif choice == 6:
        obj.front()
    elif choice == 7:
        obj.delete()
    elif choice == 8:
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```
