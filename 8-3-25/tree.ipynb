```python
# execution is left to right
# tree represented using lists

class Tree:
    def __init__(self, data):   # each node, diff values and mem hence instance var
        self.data = data
        self.children = []

    def addChild(self, child):
        self.children.append(child)

    def __str__(self,level=0):
        ret = " "*level+str(self.data)+'\n'
        for child in self.children:
            ret +=child.__str__(level+1)
        return ret

rootNode = Tree('Drink')
hot = Tree('hot')
cold = Tree('cold')
tea = Tree('tea')
coffee = Tree('coffee')
nonAlcoholic = Tree('nonAlcoholic')
alcoholic = Tree('alcoholic')

rootNode.addChild(hot)
rootNode.addChild(cold)
hot.addChild(tea)
hot.addChild(coffee)
cold.addChild(nonAlcoholic)
cold.addChild(alcoholic)

print(rootNode)
        
```

```python
# add student info
# admin add question 
# give exam
# display result
# exit

class Question:
    def __init__(self):
        self.questions = []  
        self.options = []    # List to store lists of 4 options for each question
        self.correct_answers = []  # List to store correct answers
        self.countRightAns = 0
        self.countWrongAns = 0
        self.scorePerQuestion = 5
        self.totalScore = 0
        self.fullname = None    
        self.studentId = None
    
    def studentInfo(self,fullname, studentId):
        self.studentId = studentId
        self.fullname = fullname
        print(f'Student Name: {self.fullname}\n Student ID: {self.studentId}')
        print('---------------------------------------------------------')

    def addQuestion(self, n):
        for i in range(1,n+1):
            q = input(f'Enter question {i}: ')
            self.questions.append(q)
            
            options = []
            for j in range(1,5):
                opt = input(f'Enter option {j}: ')
                options.append(opt)
            self.options.append(options)
            
            correct = int(input('Enter correct option number (1-4): '))
            self.correct_answers.append(correct)
            print('---------------------------------------------------------')

    def giveExam(self):
        self.student_answers = []
        qlen = len(self.questions)
        for i in range(qlen):
            print(f"\nQ{i+1}. {self.questions[i]}")
            for j in range(4):
                print(f"{j+1}. {self.options[i][j]}")
            while True:
                try:
                    ans = int(input("Your answer (1-4): "))
                    if 1 <= ans <= 4:
                        break
                    print("Please enter a number between 1 and 4")
                except ValueError:
                    print("Please enter a valid number")
            self.student_answers.append(ans)

    def displayResult(self):
        print("\n=== RESULTS ===")
        print(f"Student Name: {self.fullname}")
        print(f"Student ID: {self.studentId}")
        
        self.countRightAns = 0
        for i in range(len(self.questions)):
            if self.student_answers[i] == self.correct_answers[i]:
                self.countRightAns += 1
                
        self.countWrongAns = len(self.questions) - self.countRightAns
        self.totalScore = self.countRightAns * self.scorePerQuestion
        
        print(f"\nCorrect Answers: {self.countRightAns}")
        print(f"Wrong Answers: {self.countWrongAns}")
        print(f"Total Score: {self.totalScore}/{len(self.questions)*self.scorePerQuestion}")

if __name__ == '__main__':
    ob = Question()
    while True:
        print("1. Add Student Info")
        print("2. Add Questions")
        print("3. Take Exam")
        print("4. Display Result")
        print("5. Exit")
        
        ch = int(input('Enter choice: '))
        if ch == 1:
            fullname = input('Enter your fullname: ')
            studentId = input('Enter your student ID: ')
            ob.studentInfo(fullname, studentId)
        elif ch == 2:
            noOfQuestions = int(input('Enter no. of questions: '))
            ob.addQuestion(noOfQuestions)
        elif ch == 3:
            ob.giveExam()
        elif ch == 4:
            ob.displayResult()
        elif ch == 5:       
            break
        else:
            print('Invalid choice')
```
