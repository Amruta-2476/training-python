```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None  

# Taking user input for linked list creation
n = int(input("Enter the number of nodes: "))

if n <= 0:
    print("Linked List must have at least one node.")
else:
    first_value = int(input("Enter value for node 1: "))
    head_ob = Node(first_value)  # First node (head)
    current = head_ob  # Pointer to keep track of the last node

    # Creating and connecting nodes
    for i in range(2, n + 1):
        value = int(input(f"Enter value for node {i}: "))
        new_node = Node(value)
        current.next = new_node  # Link the previous node to the new one
        current = new_node  # Move to the new last node

    # Printing the Linked List
    print("\nLinked List:")
    connect_node = head_ob
    while connect_node:
        next_address = "None" if connect_node.next is None else id(connect_node.next)
        print(f'node value  [ {connect_node.data} | {next_address} ]')
        connect_node = connect_node.next
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None  

class LinkedList:
    def __init__(self):
        self.head = None  

    def append(self, data):
        """Insert a new node at the end."""
        new_node = Node(data)
        if not self.head:  
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node  

    def print_list(self):
        """Traverse and print the linked list."""
        temp = self.head
        while temp:
            print(f"[ {temp.data} | {id(temp.next)} ] ->", end=" ")
            temp = temp.next
        print("None")

# Testing the LinkedList
ll = LinkedList()
ll.append(10)
ll.append(23)
ll.append(30)

ll.print_list()
```
