```python
# stack without size
# push, pop, peek, isEmpty, isFull(not here), delete, show
class Stack:
    def __init__(self):
        self.mystack = []

    def push(self, val):
        self.mystack.append(val)
        print(f'Pushed {val} on stack.')

    def pop(self):
        if self.isEmpty():
            print('Stack empty. Cannot pop')
        else:
            popped_val = self.mystack.pop()
            # popped_val = self.mystack.remove(self.mystack[-1])
            print(f'Popped {popped_val} from stack.')

    def peek(self):
        if self.isEmpty():
            print('Stack empty. Cannot peek')
        else:
            print(f'Top element is {self.mystack[-1]}')

    def show(self):
        print(self.mystack,' is the stack.')

    def isEmpty(self):
        return len(self.mystack) == 0
    
    def delete(self):
        self.mystack = []
        # self.mystack = None     these
        # del(self.mystack)       are
        # self.mystack.clear()    different ways
        print(self.mystack)
        

ob = Stack()
while True:
    print('1. Push')
    print('2. Pop')
    print('3. Peek')
    print('4. isEmpty')
    print('5. Show')
    print('6. Delete')
    print('7. Exit')
    ch = int(input('enter choice: '))

    if ch == 1:
        val = int(input('enter value to push: '))
        ob.push(val)
    elif ch == 2:
        ob.pop()
    elif ch == 3:
        ob.peek()
    elif ch == 4:
        print(ob.isEmpty())
    elif ch == 5:
        ob.show()
    elif ch == 6:
        ob.delete()
    elif ch == 7:
        print('exiting . . .')            # or you can do sys.exit() from sys module....breaks from loop same like break
        break
    else:
        print('invalid choice')
```

```python
# stack with size
# push, pop, peek, isEmpty, isFull, delete, show
class Stack:
    def __init__(self, max_size):
        self.mystack = []
        self.max = max_size

    def push(self, val):
        if self.isFull():
            print('Stack full. Cannot push')
        else:
            self.mystack.append(val)
            print(f'Pushed {val} on stack.')

    def pop(self):
        if self.isEmpty():
            print('Stack empty. Cannot pop')
        else:
            popped_val = self.mystack.pop()
            # popped_val = self.mystack.remove(self.mystack[-1])
            print(f'Popped {popped_val} from stack.')

    def peek(self):
        print(f'Top element is {self.mystack[-1]}')

    def show(self):
        print(self.mystack,' is the stack.')

    def isEmpty(self):
        return len(self.mystack) == 0
    
    def isFull(self):
        return len(self.mystack) ==self.max
    
    def delete(self):
        self.mystack.clear()
        print(self.mystack)
        

max = int(input('enter max size of stack: '))
ob = Stack(max)
while True:
    print('1. Push')
    print('2. Pop')
    print('3. Peek')
    print('4. isEmpty')
    print('5. isEmpty')
    print('6. Show')
    print('7. Delete')
    print('8. Exit')
    ch = int(input('enter choice: '))

    if ch == 1:
        val = int(input('enter value to push: '))
        ob.push(val)
    elif ch == 2:
        ob.pop()
    elif ch == 3:
        ob.peek()
    elif ch == 4:
        print(ob.isEmpty())
    elif ch == 5:
        ob.isEmpty()
    elif ch == 6:
        ob.show()
    elif ch == 7:
        ob.delete()
    elif ch == 8:
        print('exiting . . .')
        break
    else:
        print('invalid choice')
```

```python
# TOWER OF HANOI Problem
# A =[3,2,1]   B=[]    C=[]

class Tower:
    def __init__(self):
        print("Welcome to Tower of Hanoi")
        print("Given problem: A=[3,2,1]      B=[]      C=[]")
        print("Expected output: A=[]      B=[]      C=[3,2,1]")
        self.A = []
        self.B = []
        self.C = []
        
    def move_disk(self, source, destination, source_name, dest_name):
        if len(source) == 0:
            return
        disk = source.pop()
        destination.append(disk)
        print(f"Move disk {disk} from {source_name} to {dest_name}")
        print(f"A={self.A}  B={self.B}  C={self.C}")
    
    def solve_tower(self, n, source, auxiliary, destination, 
                   source_name='A', aux_name='B', dest_name='C'):
        if n == 1:
            self.move_disk(source, destination, source_name, dest_name)
            return
        
        # Move n-1 disks from source to auxiliary
        self.solve_tower(n-1, source, destination, auxiliary, 
                        source_name, dest_name, aux_name)
        
        # Move remaining disk from source to destination
        self.move_disk(source, destination, source_name, dest_name)
        
        # Move n-1 disks from auxiliary to destination
        self.solve_tower(n-1, auxiliary, source, destination, 
                        aux_name, source_name, dest_name)

# Create instance and solve
ob = Tower()
ob.A = [3, 2, 1]  # Disks are in reverse order (largest to smallest)
ob.solve_tower(len(ob.A), ob.A, ob.B, ob.C)
```
